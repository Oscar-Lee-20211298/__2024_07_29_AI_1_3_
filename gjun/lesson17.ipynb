```python
import numpy as np
import pandas as pd

m = int(input("enter number of m"))
n = int(input("enter number of n"))
o = int(input("enter number of o"))
p = int(input("enter number of p"))
q = int(input("enter number of q"))
scores = np.random.randint(m,n,size = (m,o))

scores_df = pd.DataFrame(scores,
                    columns=['chinese', 'english', 'math', 'history','gegorgaphy'],
                    index = range(p,q))
scores_df.index.name = 'id'

scores_df.columns.name = 'subject'

scores_df

```

```python
sum_series = scores_df.sum(axis=1)
mean_series = scores_df.mean(axis=1)
rank_series = sum_series.rank(method='min', ascending=False)
scores_df['total'] = sum_series
scores_df['average'] = mean_series
scores_df['rank'] = rank_series
scores_df

```

```python
scores_sorted_df = scores_df.sort_values(by='rank',ascending=True)
scores_sorted_df
```
